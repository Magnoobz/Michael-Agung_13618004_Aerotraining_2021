```python
# Problem 1

for i in range(1,101):                           # digunakan untuk looping angka dari 1 hingga 100
    
    output = ""                                  # digunakan untuk membuat variabel string kosong
    
    if i%3 == 0:                                 # mengecek kelipatan 3
        output += "Fizz"                         # jika habis dibagi 3, output ditambah "Fizz"
    if i%5 == 0:                                 # mengecek kelipatan 5
        output += "Buzz"                         # jika habis dibagi 5, output ditambah "Buzz"
                                                 # tidak digunakan elif, shg jika habis dibagi 3 dan 5, output jadi "FizzBuzz"
    if i%3 != 0 and i%5 != 0:                    # jika tidak habis dibagi 3 maupun 5
        output = i                               # output adalah bilangan tersebut
        
    print(output)                                # mencetak output
```

```python
# Problem 2

s = input("Input\t: ").lower()                   # mengambil input user, .lower() digunakan untuk meng-ignore case

n = int(len(s)/2)                                # menghitung panjang string dibagi 2, karena palindrome simetris kanan-kiri
                                                 # jadi banyak huruf yang dicek hanya sebanyak setengah banyak huruf total

count = 0                                        # membuat variabel count
for i in range(n):                               # looping sejumlah n kali
    if s[i] == s[-(i+1)]:                        # apabila huruf ke i sama dengan ke i dari belakang, maka count ditambah 1
        count += 1                               # catatan: i dan -(i+1) karena indexing string mulai dari 0
        
if count == n:                                   # Jika count = n, semua huruf di setengah depan sama dengan setengah belakang
    print("Output\t: Yes")                       # Maka adalah palindrome
else:                                            # sebaliknya maka tidak palindrome
    print("Output\t: No")
```
