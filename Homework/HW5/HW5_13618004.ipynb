```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
data.info()
```

```python
n_train = len(data)
```

```python
data = data.append(test_data)

data
```

```python
y_train = data['Survived'][:n_train]
```

```python
data = data.drop(["PassengerId", "Survived", "Name", "Ticket", "Fare", "Embarked"], axis = 1)

data
```

```python
data = data.fillna({"Age": data["Age"].mean()})
data.info()
```

```python
data['Cabin'].unique()
```

```python
data['Cabin'] = data['Cabin'].map(lambda x: str(x)[0])

data['Cabin'].unique()
```

```python
data = pd.get_dummies(data)

data
```

```python
data = data.drop(["Cabin_n"], axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data)
scaled_data = scaler.transform(data)
scaled_data = pd.DataFrame(scaled_data, columns = data.columns)

scaled_data
```

```python
X_train = scaled_data[:n_train]
X_test = scaled_data[n_train:]
```

```python
from sklearn.svm import SVC

clf = SVC()
clf.fit(X_train, y_train)
```

```python
sample = X_train.iloc[3,:].tolist()

clf.predict([sample])
```

```python
clf.score(X_train,y_train)
```
